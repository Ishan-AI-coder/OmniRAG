```python
from typing import Dict,TypedDict    #Dict- the values should have the same data type   #TypedDict -> the keys are only string 
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):# Our StateSchema
    message:str

def greetingNode(state:AgentState)->AgentState:
    """Simple node that adds a greeeting message to the state"""   #Docstring-A special type of string   used in Python for adding more info about the modules
    state['message']= 'Hey '+state["message"]+" ,how is your day going?"
    return state
    
```

```python
graph=StateGraph(AgentState)#Creating a graph

graph.add_node("greet",greetingNode)#Adding our function as a node in the graph
graph.set_entry_point("greet")
graph.set_finish_point("greet")

app=graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result=app.invoke({"message":"Bob"})
```
