```python
from typing import TypedDict,List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values:List[int]
    name:str
    result:str
```

```python
def process_values(state:AgentState)->AgentState:
    """This function handles multiple inputs"""
    state["result"]=f'Hi there {state["name"]} .Your result is ={sum(state["values"])}'
    return state
```

```python
graph=StateGraph(AgentState)
graph.add_node("Class_results",process_values)
graph.set_entry_point("Class_results")
graph.set_finish_point("Class_results")

app=graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
answer=app.invoke({"values":[12,16,10],"name":"Bob"})
print(answer["name"])
```
